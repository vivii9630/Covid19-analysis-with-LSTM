```python
import pandas as pd
import numpy as np


df_patient = pd.read_csv(r'C:/Users/Vivek/Downloads/coviddatanew.csv',index_col="location")
df_patient = df_patient.loc["Canada"]
daily_count = df_patient.groupby(['date','total_cases']).total_cases.count()
daily_count = pd.DataFrame(daily_count)
daily_count
```

```python
data = daily_count.cumsum()
data
```

```python
dataset = data.iloc[14:]
dataset
dataset.columns = ['total_cases']
print("len of the dataset::"+str(len(dataset)))
data = np.array(dataset).reshape(-1, 1)
train_data = dataset[:len(dataset)-5]#0..35
test_data = dataset[len(dataset)-5:]#35..40

train_data
```

```python
test_data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(train_data)
scaled_train_data = scaler.transform(train_data)
scaled_test_data = scaler.transform(test_data)
scaled_train_data
```

```python

scaled_test_data
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM


from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
n_input =3
n_features =1
                             
generator = TimeseriesGenerator(scaled_train_data,scaled_train_data, length=n_input, batch_size=1)

lstm_model = Sequential()
lstm_model.add(LSTM(19, activation='relu', input_shape = (n_input, n_features)))
lstm_model.add(Dense(10))
lstm_model.add(Dense(5))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mse')
lstm_model.summary()
```

```python
lstm_model.fit_generator(generator, epochs=28)
```

```python
import matplotlib.pyplot as plt
losses_lstm = lstm_model.history.history['loss']
plt.figure(figsize = (12,4))
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.xticks(np.arange(0,21,1))
plt.plot(range(len(losses_lstm)), losses_lstm)
```

```python
lstm_predictions_scaled = []

batch = scaled_train_data[-n_input:]
current_batch = batch.reshape((1, n_input, n_features))

for i in range(len(test_data)):   
    lstm_pred = lstm_model.predict(current_batch)[0]
    lstm_predictions_scaled.append(lstm_pred) 
    current_batch = np.append(current_batch[:,1:,:],[[lstm_pred]],axis=1)


# As you know we scaled our data thats why we have to inverse it to see true predictions.
lstm_predictions_scaled
```

```python
lstm_predictions = scaler.inverse_transform(lstm_predictions_scaled) 
lstm_predictions
```

```python
test_data['LSTM_Predictions'] = lstm_predictions
test_data
```

```python
test_data.plot()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print('MAE of LSTM Model ',mean_absolute_error(test_data['total_cases'], test_data['LSTM_Predictions']))

print('MSE of LSTM Model ',mean_squared_error(test_data['total_cases'], test_data['LSTM_Predictions']))
```
